```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score,train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
train=pd.read_csv('SalaryData_Train(1).csv')
train
```

```python
train.nunique()
```

```python
columns=['workclass','education','race','maritalstatus','occupation','relationship','sex','native','Salary']
for i in columns:
  plt.show()
  print(i)
  print(sns.countplot(x=i,data=train))
```

```python
train.info()
```

```python
le=LabelEncoder()
columns=['workclass','education','race','maritalstatus','occupation','relationship','sex','native','Salary']
for i in columns:
  train[i]=le.fit_transform(train[i])
train
```

```python
x=train.iloc[:,0:13]
y=train.iloc[:,13]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=SVC(kernel='linear',C=0.01,gamma=0.1)
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred_test=model.predict(x_test)
```

```python
accur_train=accuracy_score(y_test,y_pred_test)
accur_train
```

```python
print("Accuracy of train_dataset:",accur_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_test))
```
